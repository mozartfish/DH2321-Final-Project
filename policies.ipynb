```python
import pandas as pd
```

```python
df = pd.read_csv('data/policies.csv')
df.head(2)
```

```python
df = df.rename(columns={'Name of policy or measure': 'Policy'})
df = df.rename(columns={'Implementation period start': 'Year'})
df = df.rename(columns={'Status of implementation': 'Status'})
df = df.rename(columns={'Sector(s) affected': 'Sector'})
df = df.rename(columns={'pdf': 'Source'})
df['Year'] = df['Year'].fillna(0).astype(int)


df = df[['Country', 'Year', 'Policy', 'Description', 'Status', 'Sector', 'Source']]

df.to_csv('data/policies.csv', index=False)

df
```
